```python
# import dependencies
import pandas as pd
```

```python
# read csv

eighteen_data = pd.read_csv("data/2018.csv")
twentyone_data = pd.read_csv("data/world-happiness-report-2021.csv")

og_eighteen_df = eighteen_data
og_twentyone_df = twentyone_data
```

```python
# Make copy 'eighteen_df' to rename columns
eighteen_df = og_eighteen_df.copy()

# Make copy 'twentyone_df' to rename columns
twentyone_df = og_twentyone_df.copy()
```

```python
#ADD YEAR COLUMNS TO 2018 AND 2021 DATAFRAMES

# Find length of 2018 data frame
eighteen_df_len = og_eighteen_df['Overall rank'].unique()

# Find length of 2021 data frame
twentyone_df_len = og_twentyone_df['Country name'].unique()
 

#Create an empty list 'eighteen_df_year_values' to hold
eighteen_df_year_values = []

#Create an empty list 'twentyone_df_year_values' to hold
twentyone_df_year_values = []

# Add '2018' Values to eighteen_df_year_values
for i in range(len(eighteen_df_len)):
    eighteen_df_year_values.append('2018')
    

# Add '2021' Values to twentyone_df_year_values
for i in range(len(twentyone_df_len)):
    twentyone_df_year_values.append('2021')

# Add Year Column to 'eighteen_df' data frame and name column 'Year'
eighteen_df.insert(0, "Year", eighteen_df_year_values)

# Add Year Column to 'twentyone_df' data frame and name column 'Year'
twentyone_df.insert(0, "Year", twentyone_df_year_values)
```

```python
# Rename 2018 dataframe Columns

# 'Overall rank' -> 'Happiness_Rank'
eighteen_df.rename(columns = {'Overall rank':'Happiness_Rank'}, inplace = True)

# 'Country or region' -> 'Country'
eighteen_df.rename(columns = {'Country or region':'Country'}, inplace = True)

# Copy 'Country or region' column to create a region column with the same values
# 'Country or region' -> 'Region'
eighteen_df['Region'] = eighteen_df['Country']

# 'Score' -> 'Happiness_Score'
eighteen_df.rename(columns = {'Score':'Happiness_Score'}, inplace = True)

# 'GDP per capita' -> 'Economy_GDP'
eighteen_df.rename(columns = {'GDP per capita':'Economy_GDP'}, inplace = True)

# 'Social support' -> 'Family' #UNSURE
eighteen_df.rename(columns = {'Social support':'Family'}, inplace = True)

# 'Healthy life expectancy' -> 'Life_Expectancy'
eighteen_df.rename(columns = {'Healthy life expectancy':'Life_Expectancy'}, inplace = True)

# 'Freedom to make life choices' -> 'Freedom'
eighteen_df.rename(columns = {'Freedom to make life choices':'Freedom'}, inplace = True)


# 'Perceptions of corruption' -> 'Trust_in_Gov'
eighteen_df.rename(columns = {'Perceptions of corruption':'Trust_in_Gov'}, inplace = True)
eighteen_df

# add dystopia column
eighteen_df['Dystopia_Residual']=""

# Reorder columns

```

```python
# Set 'twentyone_df' index column values as new 'Happiness_Rank' column
twentyone_df.reset_index(inplace=True)
```

```python
# # Rename 2021 dataframe Columns

# 'index' -> 'Happiness_Rank'
twentyone_df.rename(columns = {'index':'Happiness_Rank'}, inplace = True)

# 'Country name' -> 'Country'
twentyone_df.rename(columns = {'Country name':'Country'}, inplace = True)

# 'Regional indicator' -> 'Region'
twentyone_df.rename(columns = {'Regional indicator':'Region'}, inplace = True)

# 'Ladder score' -> 'Happiness_Score'
twentyone_df.rename(columns = {'Ladder score':'Happiness_Score'}, inplace = True)

# 'Logged GDP per capita' -> 'Economy_GDP'
twentyone_df.rename(columns = {'Logged GDP per capita':'Economy_GDP'}, inplace = True)

# 'Social support' -> 'Family' #UNSURE
twentyone_df.rename(columns = {'Social support':'Family'}, inplace = True)

# 'Healthy life expectancy' -> 'Life_Expectancy'
twentyone_df.rename(columns = {'Healthy life expectancy':'Life_Expectancy'}, inplace = True)

# 'Freedom to make life choices' -> 'Freedom'
twentyone_df.rename(columns = {'Freedom to make life choices':'Freedom'}, inplace = True)

# 'Perceptions of corruption' -> 'Trust_in_Gov'
twentyone_df.rename(columns = {'Perceptions of corruption':'Trust_in_Gov'}, inplace = True)

# 'Dystopia + residual' -> 'Dystopia_Residual'
twentyone_df.rename(columns = {'Dystopia + residual':'Dystopia_Residual'}, inplace = True)
```

```python
# Remove unwanted 2021 columns

# Create a list 'unwanted_columns_list' set equal to unwanted Columns in 'twentyone_df'
unwanted_columns_list = ['Standard error of ladder score','upperwhisker','lowerwhisker','Ladder score in Dystopia','Explained by: Log GDP per capita','Explained by: Social support','Explained by: Healthy life expectancy','Explained by: Freedom to make life choices','Explained by: Generosity','Explained by: Perceptions of corruption']

# Use drop() function to remove unwanted columns from 'twentyone_df'
cleaned_twentyone_df = twentyone_df.drop(unwanted_columns_list, axis=1)
cleaned_twentyone_df
```

```python
# EXPORT DATA TO CSV

# Export and 'eighteen_df' rename dataframes as csv
eighteen_df.to_csv("data/clean_2018.csv", index=False, encoding="ISO-8859-1")

# Export and 'cleaned_twentyone_df' rename dataframes as csv
cleaned_twentyone_df.to_csv("data/clean_2021.csv", index=False, encoding="ISO-8859-1")
```
