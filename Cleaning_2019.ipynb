```python
# Dependencies
import requests
import pandas as pd

# File to Load
path_19 = "data/2019.csv"

# Read the modified Comic Books csv and store into Pandas DataFrame
C19 = pd.read_csv(path_19, encoding="utf-8")

C19.head()

```

```python
# Convert to DataFrame
csv_2019_df = pd.DataFrame(C19)

# Column renaming
csv_2019_df = csv_2019_df.rename(
    columns = {
        "Country or region": "Country",
        "Overall rank": "Happiness_Rank",
        "GDP per capita": "Economy_GDP",
        "Healthy life expectancy": "Life_Expectancy",
        "Freedom to make life choices": "Freedom",
        "Perceptions of corruption": "Trust_in_Gov",
        "Score": "Happiness_Score",
        "Generosity": "Generosity"
    }
)

csv_2019_df.head()
```

```python
# Add a Year and any missing columns
csv_2019_df.loc[:,"Year"] = "2019"
csv_2019_df.loc[:,"Region"] = "nan"
csv_2019_df.loc[:,"Family"] = "nan"
csv_2019_df.loc[:,"Dystopia_Residual"] = "nan"


# Configure the final DataFrame
csv_2019_df = csv_2019_df[[
        "Country",
        "Region",
        "Happiness_Rank",
        "Happiness_Score",
        "Economy_GDP",
        "Family",
        "Life_Expectancy",
        "Freedom",
        "Trust_in_Gov",
        "Generosity",
        "Dystopia_Residual",
        "Year"
]]


# Display DataFrame length and sample data
csv_2019_df.head()
```

```python
# Save the DataFrame as a CSV
# Note: To avoid any issues later, use encoding="utf-8"
csv_2019_df.to_csv("data/csv_2019_clean.csv", encoding="utf-8", index=False)
```
